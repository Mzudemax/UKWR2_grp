### (a)
We will try to eveluate the following Integral:
$y_n=y_n(a)=\int_{0}^{1} (\frac{x^n}{x+a})dx = \frac{1}{n}-ay_{n-1}$.   
First of all we will Plot the integrand for $a=5$ and $n=1,5,10,20,30,50$ in the domain $ 0\leq x \leq 1$.

```python
x = np.arange(0.0,1,0.01) 
def func(a,n):
    return (x**n)/(x+a)

'''
This defines the Integrad dependend on a and n.
The Domain is found in the bounds of np.arange and the last argument defines the
spacing between values.

'''

def pltmulti(x,a,n):
    for i in range(1,len(n)):
        plt.plot(x,func(a,n[i-1]))
'''
This function takes an array for n and plots our function for every n in the array.
'''
plt.title("Integrand for different n") 
plt.xlabel("x-axis") 
plt.ylabel("y-axis") 
pltmulti(x,5,[1,5,10,20,30,50])
plt.show()
```

### (b) 
We now want to write a program that gets $a,n_0,y_0,n_1$ as input and then performs the iteration from $n_0$ to $n_1$.  
Therefor we first distinguish between interating up and iterating down (whether $n_1 > n_0$ or $n_1 < n_0$).  
If we iterate up we can just use the given formula to get $y_(n+1)$ from $y_n$.  
If we iterate down we have to "change" the formula: $ y_(n-1) = \frac{1}{a}(\frac{1}{n}-y_n)$  

```python
'''
The first to functions are the Integrals definition according to either iterate up or down.
'''
def int_up(a,n0,y0):
    return 1/n0 - a*y0

def int_down(a,n0,y0):
    return 1/a * (1/n0 - y0)
'''
The Iteration is performed abs(n1-n0)-1 because we start by calculating the next function after y0.
If n1-n0 is positive therefor n1 > n0 and we iterate up otherwise visa versa.
'''
def iteration(a,n0,y0,n1):
    total = n1-n0
    ni = n0
    yi = y0
    for i in range(1,abs(total)):
        if np.sign(total)>0:
            ni = ni + 1
            yi = int_up(a,ni,yi)
        else:
            ni = ni-1
            yi = int_down(a,ni,yi)
            
    return yi
        
```

### (c)
Now we should test our programm and see how it behaves.

```python
print(iteration(5,0,np.log((1+5)/5),30))  
print(iteration(5,50,1,30))
print(iteration(5,30,0.005404632965088425,50))
        
```
